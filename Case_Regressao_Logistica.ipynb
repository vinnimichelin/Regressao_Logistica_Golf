```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
#Importando o Dataset

df = pd.read_excel('golf_df.xlsx')

df.head()
```

```python
#Pegando as informações do Dataset

df.info()
```

```python
#Aplicando o LabelEncoder nas colunas

le = LabelEncoder()

le.fit(df.clima)
df.clima = le.transform(df.clima)

le.fit(df.temp)
df.temp = le.transform(df.temp)

le.fit(df.umidade)
df.umidade = le.transform(df.umidade)

le.fit(df.vento)
df.vento = le.transform(df.vento)

le.fit(df.jogar)
df.jogar = le.transform(df.jogar)

df = pd.DataFrame(df)
```

```python
df.head()
```

```python
#Separando o X e o Y

x = df.iloc[:, :-1]

y = df[['jogar']]
```

```python
#Treinando o modelo de regressão logistica

model = LogisticRegression()
model.fit(x, y)
```

```python
#Fazendo a predição

predictions = model.predict(x)
```

```python
#Inserindo a predição na tabela X

x["prediction"] = predictions
x.head(20)
```

```python
y = pd.DataFrame(y)
y.head()
```

```python
#Comparando a predição com o valor real, temos cerca de 80% de acerto das predições.

x["real"] = y["jogar"]
x.head(20)
```
